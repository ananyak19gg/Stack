```python
import numpy as np

arr = np.array([41, 42, 43, 44])

# Create an empty list
filter_arr = []

# go through each element in arr
for element in arr:
  # if the element is higher than 42, set the value to True, otherwise False:
  if element > 42:
    filter_arr.append(True)
  else:
    filter_arr.append(False)

newarr = arr[filter_arr]

print(filter_arr)
print(newarr)
```

# Numpy Random

```python
from numpy import random

x = random.randint(100)
y = random.rand()
z = random.randint(100, size=(5))
t = random.randint(100, size=(3, 5))
f = random.rand(5)

print(x)
print(y)
print(z)
print(t)
print(f)
```

```python
x = random.choice([3, 5, 7, 9])
x = random.choice([3, 5, 7, 9], size=(3, 5))

print(x)
print(y)
```

### probability density function in numpy

```python
x = random.choice([3, 5, 7, 9], p=[0.1, 0.3, 0.6, 0.0], size=(100))

print(x)
```

normal distribution
loc = mean,
scale = standard deviation,
size = shape of the returned rray

```python
from numpy import random
x = random.normal(size=(2,3))
print(x)
```

```python
from numpy import random
import matplotlib.pyplot as plt
import seaborn as sns
sns.displot(random.normal(size=1000), kind ="kde")
y = random.normal(loc = 1, scale = 2, size = (2,3))
print(y)
plt.show()
```

binomial distribution --> n=no. of trials, p= probability of each trials, size= shape of returned array

```python
from numpy import random
x = random.binomial(n=10,p=0.5,size=10)
print(x)
```

```python
from numpy import random
import matplotlib.pyplot as plt
import seaborn as sns

sns.displot(random.binomial(n=10, p=0.5, size=1000))

plt.show()
```

```python
from numpy import random
import matplotlib.pyplot as plt
import seaborn as sns

data = {
  "normal": random.normal(loc=50, scale=5, size=1000),
  "binomial": random.binomial(n=100, p=0.5, size=1000)
}

sns.displot(data, kind="kde")

plt.show()
```
